```python
#-------------------------------------------------------------------------------
# Name:        Map UI Generation 
# Purpose:      To generate Map from rasters
#
# Author:      Hari and Caleb
#
# Additional
# sources:      https://realpython.com/python-folium-web-maps-from-data/
#
# Created:     18/11/2023
# Copyright:   (c) Project_Canopy_Watch 2022
# Licence:     <Project_Canopy_Watch>
#-------------------------------------------------------------------------------

import os

import rasterio as rio
from rasterio import features
from rasterio.features import shapes
from rasterio.enums import Resampling

import geopandas as gpd
from matplotlib import pyplot as plt


def raster_vector(input_mask_folder, output_vector_folder):
    dir_values=os.listdir(input_mask_folder)

    for i in range(len(dir_values)):
        try:
            if dir_values[i].endswith(".tif"):
                print(dir_values[i])
                mask_input_path=os.path.join(input_mask_folder,dir_values[i])

                #raster to vector operation
                mask=None
                with rio.Env():
                    with rio.open(mask_input_path) as mask_image:
                        image_crs=mask_image.crs
                        print(image_crs)
                        mask_array=mask_image.read(1).astype(int)

                        result=({'properties': {'raster_val':v},'geometry':s}

                         for j, (s,v)
                         in enumerate(shapes(mask_array,mask=mask,transform=mask_image.transform)))

                raster_polygon_values=list(result) # conatins all the data, including No datavalues

                raster_polygon_values_minus_no_data=[] # to hold only values conatining the raster. No Data values not included
                # print('hello 1')


                # picking only pixels value 1 which corresponds to predicted features
                for k in range(len(raster_polygon_values)):
                    # if raster_polygon_values[k]['properties']['raster_val'] == 1:
                    raster_polygon_values_minus_no_data.append(raster_polygon_values[k])
                    # print('hello 2')


                # create a geodataframe housing the vector geometry and save to file as a geojson.
                new_gpd=gpd.GeoDataFrame.from_features(raster_polygon_values_minus_no_data)
                new_gpd=new_gpd.set_crs(image_crs)
                output_vector=dir_values[i].replace(".tif",".geojson")
                # print('hello 3')
                new_gpd.to_file(os.path.join(output_vector_folder,output_vector))
        except:pass


"""

###example
#location of the folder containing the raster masks to be vectorised
input_mask_folder=r'C:\...\mask_Habitation_0_patches'

# folder where the created vector will be stored
output_vector_folder=r'C:\...\vector'

##run the function
raster_vector(input_mask_folder, output_vector_folder)
"""
```
